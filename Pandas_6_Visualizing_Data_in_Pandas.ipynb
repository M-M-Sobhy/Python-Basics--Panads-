# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"/content/sample_data/Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
df.plot(kind='line',title='Ice Cream Ratings',xlabel='Daily Ratings',ylabel='Scores')
```

```python
df['Flavor Rating'].plot(kind='bar',stacked=True)
```

```python
df.plot.barh(stacked=True)
```

```python
plt.scatter(df['Texture Rating'],df['Overall Rating'],s=500,c='Yellow')
```

```python
df.plot.scatter(x='Texture Rating',y='Overall Rating',s=500,c='Yellow')
```

```python
df.plot.hist(bins=10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Flavor Rating',figsize=(30,20))
```
