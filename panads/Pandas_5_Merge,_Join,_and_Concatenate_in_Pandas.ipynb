# Merge, Join, and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"/content/LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"/content/LOTR 2.csv")
df2
```

```python
"""
how='inner'  Inner join  keeps only the rows where the join keys match in both DataFrames.
how='left'  Keep all rows from df1, match from df2 if possible.
how='right'  Keep all rows from df2, match from df1 if possible.
how='outer'  Keep all rows from both, fill missing with NaN.
"""
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
#lsuffix='_Left' / rsuffix='_Right'  If there are columns with the same name in both DataFrames, add these suffixes to distinguish them.
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df4
```

# Explanation of parameters

[df1, df2]  The list of DataFrames to combine.

axis=1  Concatenate side-by-side (columns).

(axis=0 would stack them vertically, like appending rows).

join='outer'  Union of the row indexes (keep all).

Missing values filled with NaN.

# Key difference from .merge() / .join()

concat does not look at columns to match  it just aligns by index.

merge / join match on key columns or indexes.

 So with concat, you need to make sure your indexes line up if you want matching behavior. Otherwise, youll get NaNs.

# Rule of thumb:

Use concat when you just want to glue DataFrames together (row-wise or column-wise).

Use merge / join when you want to do database-style joins.

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```

# What happens

axis=0 (default)  stack rows vertically.

join='outer' (default)  keep all columns from both DataFrames.

If a column doesnt exist in one DataFrame, its filled with NaN.
